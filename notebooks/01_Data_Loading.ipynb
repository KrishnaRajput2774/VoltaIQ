# VoltIQ – 01: Data Loading

**Objective**: Load and inspect all 8 canonical VoltIQ datasets.

This notebook demonstrates the `DataLoader` utility and confirms every dataset is accessible from `VoltIQ/datasets/`.

> **Important**: Do NOT create, download, or duplicate any datasets. Only the provided datasets in `VoltIQ/datasets/` are used.

## 1. Environment Setup

```python
import sys, os
# Ensure VoltIQ root is on the Python path
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.insert(0, project_root)
print('Project root:', project_root)
```

## 2. Import DataLoader

```python
import logging
logging.basicConfig(
    level=logging.INFO,
    format='[%(asctime)s] %(levelname)s %(name)s — %(message)s',
    datefmt='%H:%M:%S'
)

from app.utils.data_loader import DataLoader, DATASET_REGISTRY

loader = DataLoader()
print('Registered datasets:', list(DATASET_REGISTRY.keys()))
```

## 3. Verify Dataset File Structure

```python
status = loader.verify_dataset_structure()
print('\nDataset availability:')
for key, exists in status.items():
    icon = '✅' if exists else '❌ MISSING'
    print(f'  {icon}  {key}')
```

## 4. Load All Datasets

```python
dfs = loader.load_all()
print(f'\nLoaded {len(dfs)} datasets.')
```

## 5. Inspect Shapes & Column Counts

```python
import pandas as pd

summary = []
for key, df in dfs.items():
    summary.append({
        'Dataset': key,
        'Rows': len(df),
        'Columns': len(df.columns),
        'Memory (MB)': round(df.memory_usage(deep=True).sum() / 1024**2, 2)
    })

summary_df = pd.DataFrame(summary).set_index('Dataset')
print(summary_df.to_string())
```

## 6. Preview Each Dataset

```python
for key, df in dfs.items():
    print(f'\n{'='*60}')
    print(f'  {key}')
    print(f'{'='*60}')
    print(df.head(2).to_string())
    print(f'  Columns: {list(df.columns)[:8]} ...')
```

## 7. Load a Single Dataset by Key

```python
# Example: load only the battery dataset
battery_df = loader.load('battery')
print('Battery dataset shape:', battery_df.shape)
print(battery_df.dtypes)
```

## Summary

All datasets loaded successfully. The `DataLoader` correctly resolves paths from `VoltIQ/datasets/` and provides both bulk (`load_all()`) and individual (`load(key)`) access patterns.
